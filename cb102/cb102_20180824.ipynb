### KNN
- https://www.youtube.com/watch?v=UqYde-LULfs

```r
library(C50)
data(churn)
#
variable.list = !names(churnTrain) %in% c('state','area_code','account_length')
churnTrain=churnTrain[,variable.list]
set.seed(2)
#training data  testing data
ind<-sample(1:2, size=nrow(churnTrain), replace=T, prob=c(0.7, 0.3))
trainset=churnTrain[ind==1,]
testset=churnTrain[ind==2,]
```

```r
str(trainset)
```

```r
levels(trainset$international_plan) = c("1","2")
levels(trainset$voice_mail_plan) = c("1","2")
trainset$international_plan= as.numeric(trainset$international_plan)
trainset$voice_mail_plan= as.numeric(trainset$voice_mail_plan)
```

```r
levels(testset$international_plan) = c("1","2")
levels(testset$voice_mail_plan) = c("1","2")
testset$international_plan= as.numeric(testset$international_plan)
testset$voice_mail_plan= as.numeric(testset$voice_mail_plan)
```

```r
trainset_s = scale(trainset[,-17])
testset_s = scale(testset[,-17])
```

```r
#install.packages("class")
library(class)
```

```r
?knn
```

```r
predictions = knn(trainset_s,testset_s,trainset[,"churn"],k=3)
```

```r
table(predictions,testset[,'churn'])
```

```r
library('caret')
library('e1071')
```

```r
confusionMatrix(table(predictions,testset[,'churn']))
```
